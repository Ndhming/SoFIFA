```python
#Chu·∫©n h√≥a t√™n c·ªôt ƒë·ªÉ tr√°nh l·ªói kho·∫£ng tr·∫Øng
df.columns = df.columns.str.strip()
```

```python
# Chu·∫©n h√≥a c·ªôt Chi·ªÅu cao v√† C√¢n n·∫∑ng v·ªÅ cm v√† kg
df['Weight'] = df['Weight'].str.split(' ').str[0]
pattern = re.compile(r'\D+')
df['Weight'] = [int(pattern.sub('', weight)) for weight in df['Weight']]

df['Height'] = df['Height'].str.split(' ').str[0]
df['Height'] = [int(pattern.sub('', height)) for height in df['Height']]

df.head()
```

```python
# H√†m chuy·ªÉn ƒë·ªïi gi√° tr·ªã ti·ªÅn t·ªá th√†nh s·ªë th·ª±c
def convert_currency(value):
    if isinstance(value, str):  # Ki·ªÉm tra n·∫øu gi√° tr·ªã l√† chu·ªói
        value = value.replace("‚Ç¨", "").strip()  # Lo·∫°i b·ªè k√Ω hi·ªáu ‚Ç¨
        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000  # Nh√¢n v·ªõi 1 tri·ªáu
        elif 'K' in value:
            return float(value.replace('K', '')) * 1_000  # Nh√¢n v·ªõi 1 ngh√¨n
        else:
            try:
                return float(value)  # N·∫øu ch·ªâ l√† s·ªë, chuy·ªÉn th√†nh float
            except ValueError:
                return None  # X·ª≠ l√Ω gi√° tr·ªã kh√¥ng h·ª£p l·ªá
    return None  # Tr∆∞·ªùng h·ª£p gi√° tr·ªã kh√¥ng ph·∫£i chu·ªói

# √Åp d·ª•ng h√†m cho 3 c·ªôt ti·ªÅn t·ªá
df["Value"] = df["Value"].apply(convert_currency)
df["Wage"] = df["Wage"].apply(convert_currency)
df["Release clause"] = df["Release clause"].apply(convert_currency)

df.head()
```

```python
# H√†m chu·∫©n h√≥a d·ªØ li·ªáu s·ªë
def clean_numeric(value):
    if isinstance(value, (int, float)):
        return int(value)  # N·∫øu l√† s·ªë, tr·∫£ v·ªÅ lu√¥n

    value = str(value)  # ƒê·∫£m b·∫£o ƒë·∫ßu v√†o l√† chu·ªói
    value = re.sub(r'[^0-9+-]', '', value)  # Lo·∫°i b·ªè k√Ω t·ª± kh√¥ng mong mu·ªën
    match = re.match(r'(-?\d+)\s*([+-])\s*(\d+)', value)
    if match:
        base = int(match.group(1))
        operator = match.group(2)
        number = int(match.group(3))

        return base + number if operator == '+' else base - number

    return int(value)  # Tr·∫£ v·ªÅ s·ªë nguy√™n n·∫øu kh√¥ng c√≥ ph√©p to√°n

# X√°c ƒë·ªãnh c√°c c·ªôt s·ªë (lo·∫°i b·ªè c√°c c·ªôt kh√¥ng li√™n quan nh∆∞ t√™n c·∫ßu th·ªß, c√¢u l·∫°c b·ªô)
numeric_columns = df.columns.difference(["Name", "Team Contracts", "Preferred foot", "Best Position"])

# √Åp d·ª•ng chu·∫©n h√≥a cho t·ª´ng c·ªôt s·ªë
df[numeric_columns] = df[numeric_columns].apply(lambda col: col.map(clean_numeric))

df.head()
```

```python
# üîπ T√°ch "Name" th√†nh "Player Name" v√† "Position"
df[["Player Name", "Position"]] = df["Name"].str.extract(r"^(.*?)(?:\\n|\s)([A-Z]{2,}.*)?$")

# üîπ X√≥a c·ªôt "Name" c≈© n·∫øu kh√¥ng c·∫ßn thi·∫øt
df.drop(columns=["Name"], inplace=True)

# üîπ Chuy·ªÉn "Player Name" v√† "Position" l√™n ƒë·∫ßu b·∫£ng
new_order = ["Player Name", "Position"] + [col for col in df.columns if col not in ["Player Name", "Position"]]
df = df[new_order]  # ƒê·ªãnh nghƒ©a l·∫°i th·ª© t·ª± c·ªôt

df.head()
```
