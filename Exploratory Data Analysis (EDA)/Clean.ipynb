Chuẩn hóa dữ liệu

```python
#Chuẩn hóa tên cột để tránh lỗi khoảng trắng
df.columns = df.columns.str.strip()
```

```python
# Chuẩn hóa cột Chiều cao và Cân nặng về cm và kg
df['Weight'] = df['Weight'].str.split(' ').str[0]
pattern = re.compile(r'\D+')
df['Weight'] = [int(pattern.sub('', weight)) for weight in df['Weight']]

df['Height'] = df['Height'].str.split(' ').str[0]
df['Height'] = [int(pattern.sub('', height)) for height in df['Height']]

df.head()
```

```python
# Hàm chuyển đổi giá trị tiền tệ thành số thực
def convert_currency(value):
    if isinstance(value, str):  # Kiểm tra nếu giá trị là chuỗi
        value = value.replace("€", "").strip()  # Loại bỏ ký hiệu €
        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000  # Nhân với 1 triệu
        elif 'K' in value:
            return float(value.replace('K', '')) * 1_000  # Nhân với 1 nghìn
        else:
            try:
                return float(value)  # Nếu chỉ là số, chuyển thành float
            except ValueError:
                return None  # Xử lý giá trị không hợp lệ
    return None  # Trường hợp giá trị không phải chuỗi

# Áp dụng hàm cho 3 cột tiền tệ
df["Value"] = df["Value"].apply(convert_currency)
df["Wage"] = df["Wage"].apply(convert_currency)
df["Release clause"] = df["Release clause"].apply(convert_currency)

df.head()
```

```python
# Hàm chuẩn hóa dữ liệu số
def clean_numeric(value):
    if isinstance(value, (int, float)):
        return int(value)  # Nếu là số, trả về luôn

    value = str(value)  # Đảm bảo đầu vào là chuỗi
    value = re.sub(r'[^0-9+-]', '', value)  # Loại bỏ ký tự không mong muốn
    match = re.match(r'(-?\d+)\s*([+-])\s*(\d+)', value)
    if match:
        base = int(match.group(1))
        operator = match.group(2)
        number = int(match.group(3))

        return base + number if operator == '+' else base - number

    return int(value)  # Trả về số nguyên nếu không có phép toán

# Xác định các cột số (loại bỏ các cột không liên quan như tên cầu thủ, câu lạc bộ)
numeric_columns = df.columns.difference(["Name", "Team Contracts", "Preferred foot", "Best Position"])

# Áp dụng chuẩn hóa cho từng cột số
df[numeric_columns] = df[numeric_columns].apply(lambda col: col.map(clean_numeric))

df.head()
```

```python
# 🔹 Tách "Name" thành "Player Name" và "Position"
df[["Player Name", "Position"]] = df["Name"].str.extract(r"^(.*?)(?:\\n|\s)([A-Z]{2,}.*)?$")

# 🔹 Xóa cột "Name" cũ nếu không cần thiết
df.drop(columns=["Name"], inplace=True)

# 🔹 Chuyển "Player Name" và "Position" lên đầu bảng
new_order = ["Player Name", "Position"] + [col for col in df.columns if col not in ["Player Name", "Position"]]
df = df[new_order]  # Định nghĩa lại thứ tự cột

df.head()
```
